# Calculer des statistiques de prix

```python
# l'obectif ici est de calculer la moyenne des prix des billets en foction de la categorie(1,2,3)
#1Ã¨re mÃ©thode
fare_by_class = {}
passenger_classes = [1,2,3]

for this_class in passenger_classes:
    pclass_rows = titanic_survival[titanic_survival["pclass"] == this_class]
    pclass_fares = pclass_rows["fare"]
    fare_for_class = pclass_fares.mean()
    fare_by_class[this_class] = fare_for_class
 #pclass_rows
print(fare_by_class)
```

```python
#2Ã¨me methode dataframe.pivot_table()
import numpy 
passenger_class_fare = titanic_survival.pivot_table(index="pclass", values="fare", aggfunc=numpy.mean) 
passenger_class_fare
```

# appliquer des fonctions sur un dataframe

```python
#dataframe.apply(). par defaut la fonction s'applique sur toute les colonnes.
#soit une fonction qui retourne le 100e element de chaque colonne 

def row_100(column):
    item = column.iloc[99]
    return item 

row_100_var = titanic_survival.apply(row_100)
row_100_var
```

```python
#fonction qui retourne le nombre d'element nul de chaque colonne du dataframe
def null_count(column):
    column_null = pd.isnull(column)
    null = column[column_null]
    return len(null)

null = titanic_survival.apply(null_count)
null
```

```python
# appliquer une foncion a une ligne: dataframe.apply(function, axis=1)
def which_class(row):
    pclass = row["pclass"]
    
    if pd.isnull(pclass):
        return "unknow"
    elif pclass == 1:
        return "first class"
    elif pclass == 2:
        return "second class"
    else:
        return "third class"

name_classes = titanic_survival.apply(which_class, axis=1)
name_classes    

    
```

```python
def generate_age_label(row):
    age = row["age"]
    if pd.isnull(age):
        return "unknow"
    elif age < 18: 
        return "mineur"
    else:
        return "adult"
    
age_labels = titanic_survival.apply(generate_age_label, axis=1)
age_labels
```

```python
titanic_survival["age_labels"] =  age_labels
age_group_survival = titanic_survival.pivot_table(index="age_labels", values = "survived", aggfunc=numpy.mean)
age_group_survival
```
