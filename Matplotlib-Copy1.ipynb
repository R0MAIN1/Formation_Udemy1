```python
# pandas.series.dt 
import pandas as pd 
df = pd.read_csv("unrate.csv")
df['DATE'] = pd.to_datetime(df['DATE'])
df['MONTH'] = df['DATE'].dt.month
df
```

```python
fig = plt.figure(figsize= (6,6))

plt.plot(df[0:12]['MONTH'], df[0:12]['UNRATE'], c= 'red')
plt.plot(df[12:24]['MONTH'], df[12:24]['UNRATE'], c= 'green')
                  
plt.show()
```

```python
# Exercice 
fig = plt.figure(figsize= (6,3))

plt.plot(df[0:12]['MONTH'], df[0:12]['UNRATE'], c="red")
plt.plot(df[12:24]['MONTH'], df[12:24]['UNRATE'], c="blue")

plt.show()
```

```python
fig = plt.figure(figsize = (12,12))
color = ['red','blue','green','orange','black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = df[start_index:end_index]
    
    plt.plot(subset['MONTH'], subset['UNRATE'], c = color[i])
    
plt.show()
```

```python
plt.plot(df[0:12]['MONTH'], df[0:12]['UNRATE'], c='red', label='1948')
plt.plot(df[12:24]['MONTH'], df[12:24]['UNRATE'], c='red', label='1949')
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure(figsize = (12,12))
color = ['red','blue','green','orange','black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = df[start_index:end_index]
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['UNRATE'], c = color[i], label = label)
    
    plt.legend(loc="upper left")
plt.show()
```

# COURBE FINALE

```python
fig = plt.figure(figsize = (12,12))
color = ['red','blue','green','orange','black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = df[start_index:end_index]
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['UNRATE'], c = color[i], label = label)
    plt.legend(loc="upper left")
    plt.title("Evolution du taux de chomage de 1948 a 1952")
    plt.xlabel("Mois (de 1 a 12)")
    plt.ylabel("Taux de chomage en %")
plt.show()
```
